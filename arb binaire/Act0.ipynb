# Activité 3

Ecrire une fonction `nombre_de_noeuds(arbre)` qui prend en argument un objet de type `Noeud` et qui renvoie le **nombre de noeuds** contenus dans `arbre` (à gauche ou à droite)!

Exemple  :  

    >>> nombre_de_noeuds(Noeud( Noeud(None,'A',None), 'racine', Noeud(None,'B',None) ))
    3
    >>> nombre_de_noeuds(Noeud(None,'A',None) )
    1


```python
class Noeud:
    def __init__(self, g, v, d):
        self.gauche = g
        self.valeur = v        
        self.droit  = d

def nombre_de_noeuds(arbre):
    if arbre is None : 
        return 0
    # else :
    cpt = 1
    cpt += nombre_de_noeuds(arbre.droit)
    cpt += nombre_de_noeuds(arbre.gauche)
    return cpt 
```

# Activité 4

Ecrire une fonction `profondeur(arbre)` qui prend en argument un objet de type `Noeud` et qui renvoie la profondeur de `arbre`.

Exemple  :  

    >>> profondeur(Noeud( Noeud(None,'A',None), 'racine', Noeud(None,'B',None) ))
    2
    >>> profondeur(Noeud(None,'A',None) )
    1


```python
def profondeur1(arbre):
    return 0 if arbre is None else 1 + profondeur1(arbre.droit) if profondeur1(arbre.droit) > profondeur1(arbre.gauche) else 1 + profondeur1(arbre.gauche)
```

```python
def profondeur2(arbre):
    return 0 if arbre is None else 1 + max(profondeur2(arbre.droit), profondeur1(arbre.gauche))
```

```python
profondeur1(Noeud( Noeud(None,'A',None), 'racine', Noeud(Noeud(None,'C', None),'B',None) ))
```

```python
profondeur2(Noeud( Noeud(None,'A',None), 'racine', Noeud(Noeud(None,'C', None),'B',None) ))
```
