## Ecrire une fonction

Ecrire une fonction `recherche(abr, x)` qui 
* prend en arguments un arbre binaire de recherche `abr` et une valeur `x`
* renvoie `True` si `x` est l'une des valeurs contenues dans `abr`
* renvoie `False` sinon

On suggère d'écrire cette fonction de manière récursive. 

```python
def recherche(abr, x):
    if abr.racine is None:
        return False

    if x < abr.racine.valeur :
        return recherche(abr.racine.gauche, x)
    
    elif x > abr.racine.valeur :
        return recherche(abr.racine.droite, x)
    
    elif x == abr.racine.valeur:
        return True
```

```python
# pour tester le code
a = ABR()
a.insere(100)
a.insere(50)
a.insere(40)
a.insere(60)
print( recherche(a,40))  # True
print( recherche(a,0))   # False
```
