```python
import hashlib
import uuid
import platform

def generate_key():
    hasher = hashlib.sha256()
    try:
        mac_address = uuid.UUID(int=uuid.getnode()).hex[-12:]
        hasher.update(mac_address.encode('utf-8'))
    except AttributeError:
        print("No MAC address found.")

    return hasher.hexdigest()

def generate_key2(seed):
    hasher = hashlib.sha512()
    hasher.update(seed.encode('utf-8'))
    return hasher.hexdigest()

# Example usage:
key1 = generate_key()
key2 = generate_key2("your_seed_string")

print(f"Generated Key 1: {key1}")
print(f"Generated Key 2: {key2}")
```
