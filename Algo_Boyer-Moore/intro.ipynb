```
def recherche(gene, seq_adn):
    n = len(seq_adn)
    g = len(gene)
    i = 0
    while i < n-g+1 :
        j = 0
        car_gene = gene[j]
        car_adn = seq_adn[i+j]
        while j < g and car_gene == car_adn:
            j = j+1
            if j == g:
                return True
            car_gene = gene[j]
            car_adn = seq_adn[i+j]
        
        i = i+1
    return False

print(recherche("AGTC", "GTAGCAGTCTCGAAGTCGA"))
```
